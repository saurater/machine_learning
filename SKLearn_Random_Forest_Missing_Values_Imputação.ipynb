**dropna** - Dropa Valores Nulos

**axis** - Opcional, padrão 0.


*   0 e 'index' remove **Linhas** que contém valores Nulos
*   1 e 'columns' remove **Colunas** que contém valores Nulos

# Estratégia

Como há relativamente algumas entradas ausentes nos dados (tem coluna com quase 20%), podemos esperar que a eliminação de colunas não produza bons resultados. Isso ocorre porque estaríamos jogando fora muitos dados valiosos e, portanto, a imputação provavelmente terá um desempenho melhor.

# Função para Comparar Diferentes Técnicas
Para comparar Diferentes Técncias para lidar com valores ausentes, vamos criar uma função chamada escore_do_dataset. Esta função relata o **Erro Absoluto Médio (MAE)** de um modelo de **Random Florest**.

**n_estimators** -  Este é o número de árvores que você deseja construir antes de obter a votação máxima ou as médias das previsões. Um número maior de árvores oferece melhor desempenho, mas torna seu código mais lento. Você deve escolher o valor mais alto que seu processador possa suportar, pois isso torna suas previsões mais fortes e estáveis.

**fit** -  Ajusta o modelo. Em outras palavras, treina o modelo, recebendo, por parâmetro, as **preditoras** e a **alvo**.

**random_state** - 
Esse parâmetro facilita a replicação de uma solução. Um valor definido de random_state sempre produzirá os mesmos resultados se fornecido com os mesmos parâmetros e dados de treinamento.

```python
def escore_do_dataset(X_treino, X_validacao, y_treino, y_validacao):
    
    meu_modelo_random_forest = RandomForestRegressor(n_estimators=100, random_state=0)
    meu_modelo_random_forest.fit(X_treino, y_treino)
    preds = meu_modelo_random_forest.predict(X_validacao)
    
    return mean_absolute_error(y_validacao, preds)
```

```python
print("MAE (De Colunas com Valores Ausentes Dropadas):")
mae_colunas_dropadas = round(escore_do_dataset(reduzido_X_treino, reduzido_X_validacao, y_treino, y_validacao))
print(mae_colunas_dropadas)
```

# Imputação
Transformador - Imputa valores ausentes com a média, media, mais comum ou constante, tendo a média como padrão

**fit** = aprende a regra

**transform** - aplica o que aprendeu com o **fit**

**fit_transform** aprende e aplica 

Chamamos o método **fit_transform()** em nossos dados de **Treinamento** 

e o método **transform()** em nossos dados de **Teste**

```python
meu_imputador = SimpleImputer()

imputado_X_treino    = pd.DataFrame(meu_imputador.fit_transform(X_treino))
imputado_X_validacao = pd.DataFrame(meu_imputador.transform(X_validacao))

# A imputação remove os nomes das colunas. Vamos colocá-las de volta
imputado_X_treino.columns    = X_treino.columns
imputado_X_validacao.columns = X_validacao.columns
```

# Parte A

Use a próxima célula de código para pré-processar os dados de treinamento e validação. 

Defina os DataFrames pré-processados ​​para final_X_treino e final_X_validacao. 

```python
# Imputatação
final_meu_imputador     = SimpleImputer(strategy='median')
final_X_treino    = pd.DataFrame(final_meu_imputador.fit_transform(X_treino))
final_X_validacao = pd.DataFrame(final_meu_imputador.transform(X_validacao))

# A imputação removeu os nomes das colunas; coloque-os de volta
final_X_treino.columns     = X_treino.columns
final_X_validacao.columns  = X_validacao.columns

X_preditoras_teste.columns = X_treino.columns
```

Execute a próxima célula de código para treinar e avaliar um modelo de Floresta Aleatória. (Observe que não usamos a função escore_do_dataset() acima, porque em breve usaremos o modelo treinado para gerar previsões de teste!)

```python
# Definir e Ajustar o Modelo
modelo = RandomForestRegressor(n_estimators=100, random_state=0)

modelo.fit(final_X_treino, y_treino)

# Obter previsões de validação e MAE
y_alvo_previsao_validacao = modelo.predict(final_X_validacao)
print("MAE (Sua Técnica):")
print(round(mean_absolute_error(y_validacao, y_alvo_previsao_validacao),2))
```
